## Base de datos de quejas del consumidor

* Vamos hacer las transformaciones necesarias para la limpieza de la base de datos

```python
# Vamos a importar las librerias
import pandas as pd
import numpy as np
import re
import seaborn as sns
import datetime
from datetime import datetime, date
```

```python
#Vamos a leer los archivos
path = '../BD/complaints.csv'
data = pd.read_csv(path, encoding= 'unicode_escape')
```

```python
#Vamos a completar la informacion nula llenando completandolos con la palabra "sin informacion"
data = data.fillna("without information")
```

```python
#Limpiamos las variables que contiene numeros aparte para no perder la informacion
variables = ['Date received','ZIP code','Date sent to company','Complaint ID']
for i in variables:
    # print(i)
    data[i] = data[i].astype(str) # Convertimos la informacion a texto
    data[i] = data[i].apply(str.lower) #Convertimos la informacion minusculas
```

```python
#Vamos a quitar los espacios de las variables que vamos a utilizar en el analisi
variables = ['Product', 'Sub-product', 'Issue', 'Sub-issue',
       'Consumer complaint narrative', 'Company public response', 'Company',
       'State', 'Tags', 'Consumer consent provided?',
       'Submitted via',  'Company response to consumer',
       'Timely response?', 'Consumer disputed?']
for i in variables:
    # print(i)
    data[i] = data[i].astype(str) # Convertimos la informacion a texto
    data[i] = data[i].apply(str.lower) #Convertimos la informacion minusculas
    data[i] = data[i].replace('\n', '', regex=True) #Vamos a quitar los saltos de pagina
    data[i] = data[i].apply(lambda x: re.sub(r"[\W\d_]+"," ",x)) #Vamos quedarnos solo con palabras
    data[i] = data[i].apply(lambda x: x.strip()) #Quitamos los espacios de los extremos de los textos
```

```python
#Vamos a dar el formato de fecha a la columna de Date received y a la de Date sent to company
data['Date received'] = pd.to_datetime(data['Date received']).apply(lambda x : x.strftime('%Y-%m-%d'))
data['Date sent to company'] = pd.to_datetime(data['Date sent to company']).apply(lambda x : x.strftime('%Y-%m-%d'))
```

```python
#Vamos a renombrar las columnas con caracteres especiales
data = data.rename(columns={'Sub-product':'Sub_product','Sub-issue':'Sub_issue','Consumer consent provided?':'Consumer consent provided','Timely response?':'Timely response','Consumer disputed?':'Consumer disputed'})
```

```python
data.to_csv("../Output/abt.csv", sep = ';', index = False)
```
